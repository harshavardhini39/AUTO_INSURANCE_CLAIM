```python
# app.py

from flask import Flask, render_template, request
import joblib

app = Flask(__name__)

# Load the pre-trained model
model = joblib.load('gb_model.pkl')

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    if request.method == 'POST':
        # Get user inputs from the HTML form
        inception_date = request.form['inception_date']
        policy_number = request.form['policy_number']
        policy_type = request.form['policy_type']
        policy_premium = float(request.form['policy_premium'])
        channel = request.form['channel']
        state = request.form['state']
        age = int(request.form['age'])
        gender = request.form['gender']
        marital_status = request.form['marital_status']
        education = request.form['education']
        profession = request.form['profession']
        driving_exp = int(request.form['driving_exp'])
        vehicle_usage = request.form['vehicle_usage']
        annual_miles = int(request.form['annual_miles'])
        previous_citations = int(request.form['previous_citations'])
        previous_accidents = int(request.form['previous_accidents'])
        coverage_type = request.form['coverage_type']
        umbrella_policy = request.form['umbrella_policy']
        accident_date = request.form['accident_date']
        police_file = request.form['police_file']
        any_eye_witness = request.form['any_eye_witness']
        hired_attorney = request.form['hired_attorney']
        make = request.form['make']
        vehicle_cost = float(request.form['vehicle_cost'])
        road_type = request.form['road_type']
        accident_severity = request.form['accident_severity']

        # Prepare the input data for prediction
        input_data = [[
            inception_date, policy_number, policy_type, policy_premium, channel, state,
            age, gender, marital_status, education, profession, driving_exp,
            vehicle_usage, annual_miles, previous_citations, previous_accidents,
            coverage_type, umbrella_policy, accident_date, police_file, any_eye_witness,
            hired_attorney, make, vehicle_cost, road_type, accident_severity
        ]]

        # Make a prediction using the loaded model
        prediction = model.predict(input_data)

        # Display the prediction result
        return render_template('index.html', prediction=prediction[0])

if __name__ == '__main__':
     app.run(debug=True,use_reloader=False)
```
